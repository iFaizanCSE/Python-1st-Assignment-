1. What is hypothesis testing in statistics?
 - Hypothesis testing is a method used to make decisions based on data. It helps determine whether a certain assumption (hypothesis) about a population is likely to be true.
2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
 - Null Hypothesis (H0H_0): Assumes there is no effect or no difference
  - Alternative Hypothesis (H1H_1): Suggests there is an effect or a difference
3. What is the significance level in hypothesis testing, and why is it important?
- The significance level (α\alpha) is the probability of rejecting the null hypothesis when it is actually true. A common value is 0.05 (5%), meaning there’s a 5% risk of making a wrong conclusion.
4. What does a P-value represent in hypothesis testing?
- The P-value tells us how likely we are to see the observed results if the null hypothesis is true. A small P-value (<0.05< 0.05) suggests strong evidence against H0H_0, meaning we should reject it.

5. How do you interpret the P-value in hypothesis testing?
 - P-value <0.05< 0.05 → Strong evidence against H0H_0 (reject it).
 - 	P-value >0.05> 0.05 → Weak evidence against H0H_0 (fail to reject it).
 - 	P-value close to 0 → Highly unlikely result under H0H_0.
 - P-value close to 1 → Very likely under H0H_0.
6. What are Type 1 and Type 2 errors in hypothesis testing?
 - Type 1 Error (False Positive): Rejecting H0H_0 when it is actually true.
 - Type 2 Error (False Negative): Failing to reject H0H_0 when it is actually false.
7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
 - One-tailed test: Tests for an effect in only one direction (e.g., "Method A is better than Method B").
  - Two-tailed test: Tests for an effect in both directions (e.g., "Method A and Method B are different, but we don’t know which is better").
8. What is the Z-test, and when is it used in hypothesis testing?
  - A Z-test is used when:
   - The sample size is large (n>30n > 30).
  - The population variance is known.
9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
 - A Z-score tells us how far a value is from the mean in terms of standard deviations.
10. What is the T-distribution, and when should it be used instead of the normal distribution?
 - A T-distribution is used when:
 - The sample size is small (n<30n < 30).
  - The population variance is unknown.
11. What is the difference between a Z-test and a T-test?
  - Z-test → Large samples (n>30n > 30), known variance.
  - T-test → Small samples (n<30n < 30), unknown variance.
12. What is the T-test, and how is it used in hypothesis testing?
 - A T-test compares means between groups to check if the difference is significant.
  - One-sample T-test: Compares a sample mean to a known value.
  - Two-sample T-test: Compares means of two groups.
13. What is the relationship between Z-test and T-test in hypothesis testing?
  - Both compare sample means to a population mean.
  - Z-test is used for large samples, while T-test is for small samples.
14. What is a confidence interval, and how is it used to interpret statistical results?
 - A confidence interval (CI) gives a range where the true population parameter is likely to be found.
15. What is the margin of error, and how does it affect the confidence interval?
 - The margin of error is the amount of uncertainty in a confidence interval. A larger sample size reduces the margin of error, making the interval narrower.
16. How is Bayes' Theorem used in statistics, and what is its significance?
 - Bayes' Theorem helps update probabilities based on new information.
17. What is the Chi-square distribution, and when is it used?
 - The Chi-square distribution is used for categorical data to test independence and goodness of fit.
18. What is the Chi-square goodness of fit test, and how is it applied?
 - It checks if a sample follows a given distribution.
19. What is the F-distribution, and when is it used in hypothesis testing?
 - The F-distribution is used to compare variances between two or more groups, mainly in ANOVA tests.
20. What is an ANOVA test, and what are its assumptions?
 - ANOVA (Analysis of Variance) compares means of three or more groups.
   - Assumptions:
     - The data is normally distributed.
     - The samples are independent.
      -  The groups have equal variances.
21. What are the different types of ANOVA tests?
 - One-way ANOVA: Compares means of one factor (e.g., test scores of students in three schools).
 - Two-way ANOVA: Compares means of two factors (e.g., effect of study hours and diet on test scores).

interpret the results

```python
from statsmodels.stats.weightstats import ztest
import numpy as np

# Sample data (heights of 10 students)
sample_data = [51, 52, 49, 50, 48, 52, 53, 47, 50, 49]

# Known population mean
population_mean = 50

# Perform Z-test
z_stat, p_value = ztest(sample_data, value=population_mean)

# Print results
print(f"Z-Statistic: {z_stat}")
print(f"P-Value: {p_value}")

# Interpret the result
alpha = 0.05  # Significance level (5%)
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference found.")
```

```python
def visualize_errors():
    x = np.linspace(-4, 4, 100)
    y = stats.norm.pdf(x, 0, 1)

    plt.plot(x, y, label="Null Distribution")

    # Type 1 error region
    plt.fill_between(x, y, where=(x > 1.96), color='red', alpha=0.3, label="Type 1 Error")

    # Type 2 error region
    plt.fill_between(x, y, where=(x < -1.96), color='blue', alpha=0.3, label="Type 2 Error")

    plt.legend()
    plt.show()

visualize_errors()
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
import scipy.stats as stats

def calculate_margin_of_error(data, confidence=0.95):
    n = len(data)  # Sample size
    mean = np.mean(data)  # Sample mean
    std_dev = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for sample)

    # Get Z-score for given confidence level
    alpha = 1 - confidence
    z_critical = stats.norm.ppf(1 - alpha/2)  # Two-tailed test

    # Compute Margin of Error
    margin_of_error = z_critical * (std_dev / np.sqrt(n))

    return margin_of_error

# Example usage
sample_data = np.random.normal(50, 5, 30)  # Generate random sample data
confidence_level = 0.95  # 95% confidence

me = calculate_margin_of_error(sample_data, confidence_level)
print(f"Margin of Error (95% confidence): {me:.4f}")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior, likelihood, false_positive):
    # Total probability of testing positive (law of total probability)
    evidence = (likelihood * prior) + (false_positive * (1 - prior))

    # Applying Bayes' Theorem
    posterior = (likelihood * prior) / evidence
    return posterior

# Given probabilities
P_D = 0.01      # Prior: 1% of people have the disease
P_T_given_D = 0.9  # Likelihood: Test detects disease correctly 90% of the time
P_T_given_not_D = 0.05  # False Positive Rate: 5%

# Compute the probability of having the disease given a positive test
posterior_prob = bayes_theorem(P_D, P_T_given_D, P_T_given_not_D)

print(f"Probability of having the disease given a positive test: {posterior_prob:.4f}")
```

16.  Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
import scipy.stats as stats

# Generate two random samples
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(50, 10, 30)  # Mean=50, Std=10, Size=30
sample2 = np.random.normal(55, 15, 30)  # Mean=55, Std=15, Size=30

# Calculate sample variances
var1 = np.var(sample1, ddof=1)  # Unbiased variance (ddof=1)
var2 = np.var(sample2, ddof=1)

# Compute F-statistic
F = var1 / var2 if var1 > var2 else var2 / var1  # Ensure F >= 1

# Compute p-value (two-tailed test)
df1 = len(sample1) - 1
df2 = len(sample2) - 1
p_value = 2 * (1 - stats.f.cdf(F, df1, df2))  # Two-tailed test

# Print results
print(f"Sample 1 Variance: {var1:.2f}")
print(f"Sample 2 Variance: {var2:.2f}")
print(f"F-statistic: {F:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA .

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(*groups):

    print("Checking ANOVA Assumptions...\n")

    # Normality Check: Shapiro-Wilk Test
    print("1️⃣ Normality Test (Shapiro-Wilk Test):")
    for i, group in enumerate(groups, start=1):
        stat, p = stats.shapiro(group)
        print(f"   Group {i}: W-statistic = {stat:.3f}, P-value = {p:.4f}")
        if p < 0.05:
            print("   ⚠️ Group does not follow normal distribution (p < 0.05)")
        else:
            print("   ✅ Group follows normal distribution (p > 0.05)")
    print("\n")

    # Homogeneity of Variance Check: Levene's Test
    print("2️⃣ Homogeneity of Variance Test (Levene’s Test):")
    stat, p = stats.levene(*groups)
    print(f"   Levene’s Test: W-statistic = {stat:.3f}, P-value = {p:.4f}")
    if p < 0.05:
        print("   ⚠️ Groups have significantly different variances (p < 0.05)")
    else:
        print("   ✅ Variances are equal across groups (p > 0.05)")
    print("\n")

    # Visualization - Boxplot to check variance visually
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=groups)
    plt.title("Boxplot of Groups for Variance Check")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.grid()
    plt.show()

# ---- Example Usage ----
np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)
group3 = np.random.normal(60, 10, 30)

check_anova_assumptions(group1, group2, group3)
```

20.  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
results.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt

# Create synthetic dataset
np.random.seed(42)

# Define factors
teaching_method = np.repeat(['Traditional', 'Online'], 30)  # Two teaching methods
study_time = np.tile(['Low', 'Medium', 'High'], 20)  # Three study time levels

# Generate exam scores (dependent variable) with some variation
scores = np.random.normal(70, 5, 60)  # Mean = 70, Std = 5

# Create DataFrame
df = pd.DataFrame({'Teaching_Method': teaching_method, 'Study_Time': study_time, 'Score': scores})

# ---- Perform Two-Way ANOVA ----
model = smf.ols('Score ~ C(Teaching_Method) + C(Study_Time) + C(Teaching_Method):C(Study_Time)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print ANOVA table
print("\n🔹 Two-Way ANOVA Results:")
print(anova_table)

# ---- Visualization ----
plt.figure(figsize=(8, 5))
sns.boxplot(x='Study_Time', y='Score', hue='Teaching_Method', data=df, palette='Set2')
plt.title("Two-Way ANOVA: Exam Scores by Teaching Method & Study Time")
plt.xlabel("Study Time")
plt.ylabel("Exam Score")
plt.legend(title="Teaching Method")
plt.grid()
plt.show()
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
import scipy.stats as stats

# Generate random sample
np.random.seed(42)
sample = np.random.normal(50, 10, 30)  # Mean=50, Std=10, Size=30

# Hypothesized population variance
sigma_squared_0 = 100  # (Since Std=10, Variance = 10^2 = 100)

# Sample statistics
n = len(sample)
sample_variance = np.var(sample, ddof=1)  # Sample variance with Bessel's correction

# Chi-square statistic
chi_square_stat = (n - 1) * sample_variance / sigma_squared_0

# Critical values (for 95% confidence level)
alpha = 0.05  # Significance level
chi2_lower = stats.chi2.ppf(alpha / 2, df=n - 1)  # Lower critical value
chi2_upper = stats.chi2.ppf(1 - alpha / 2, df=n - 1)  # Upper critical value

# Print results
print("Chi-Square Statistic:", chi_square_stat)
print("Lower Critical Value:", chi2_lower)
print("Upper Critical Value:", chi2_upper)

# Interpretation
if chi_square_stat < chi2_lower or chi_square_stat > chi2_upper:
    print("Reject the null hypothesis: Population variance is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variance.")
```

25.  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
import scipy.stats as stats

# Sample data
x1, n1 = 50, 200  # Group 1: 50 successes out of 200
x2, n2 = 30, 180  # Group 2: 30 successes out of 180

# Calculate sample proportions
p1 = x1 / n1
p2 = x2 / n2

# Pooled proportion
p = (x1 + x2) / (n1 + n2)

# Compute Z-score
z_stat = (p1 - p2) / np.sqrt(p * (1 - p) * (1/n1 + 1/n2))

# Compute p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed test

# Print results
print("Z-Statistic:", z_stat)
print("P-Value:", p_value)

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between proportions.")
```

26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results .

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate two random samples
np.random.seed(42)
sample1 = np.random.normal(50, 10, 30)  # Mean=50, Std=10, Size=30
sample2 = np.random.normal(52, 15, 30)  # Mean=52, Std=15, Size=30

# Compute sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Compute F-statistic
F = var1 / var2 if var1 > var2 else var2 / var1  # Always put larger variance in numerator
df1, df2 = len(sample1) - 1, len(sample2) - 1  # Degrees of freedom

# Compute p-value
p_value = 2 * (1 - stats.f.cdf(F, df1, df2))  # Two-tailed test

# Print results
print("F-Statistic:", F)
print("P-Value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")

# Visualization
plt.figure(figsize=(8,5))
plt.hist(sample1, bins=10, alpha=0.6, color='blue', label="Sample 1")
plt.hist(sample2, bins=10, alpha=0.6, color='red', label="Sample 2")
plt.axvline(np.mean(sample1), color='blue', linestyle='dashed', linewidth=2, label="Mean 1")
plt.axvline(np.mean(sample2), color='red', linestyle='dashed', linewidth=2, label="Mean 2")
plt.legend()
plt.title("Histogram of Two Samples")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results .

```python
import numpy as np
import scipy.stats as stats

# Simulated observed data (e.g., frequencies of categories)
observed = np.array([50, 30, 20])  # Observed counts in three categories

# Expected proportions (assume uniform distribution)
total = np.sum(observed)
expected = np.array([total / 3] * 3)  # Equal expected frequencies

# Perform Chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(observed, expected)

# Print results
print("Chi-Square Statistic:", chi2_stat)
print("P-Value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")
```
