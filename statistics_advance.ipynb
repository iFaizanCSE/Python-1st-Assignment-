1. What is a random variable in probability theory?
- A random variable is a function that assigns numerical values to outcomes of a random experiment.
2. What are the types of random variables?
- Discrete Random Variables (countable outcomes, e.g., rolling a die)
Continuous Random Variables (infinite possible values, e.g., height of people)
3. What is the difference between discrete and continuous distributions?
- Discrete distribution deals with countable values (e.g., binomial distribution).
- Continuous distribution deals with uncountable values (e.g., normal distribution).
4. What are probability distribution functions (PDF)?
- A function that describes the likelihood of a random variable taking certain values.
5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
 - PDF gives the probability density for a specific value.
 - CDF gives the cumulative probability up to a given value.
6. What is a discrete uniform distribution?
- A distribution where each outcome has an equal probability (e.g., rolling a fair die).
7. What are the key properties of a Bernoulli distribution?
= Two possible outcomes: success (1) and failure (0).

8. What is the binomial distribution, and how is it used in probability?

9. What is the Poisson distribution and where is it applied?
- Models the number of events in a fixed interval of time/space.
- Used in call centers, traffic flow, etc.
10. What is a continuous uniform distribution?
-  A distribution where all values in a range are equally likely (e.g., choosing a random number between 0 and 1).
11. What are the characteristics of a normal distribution?
- Bell-shaped curve.
- Symmetrical around the mean.
- Defined by mean (μ) and standard deviation (σ).
12. What is the standard normal distribution, and why is it important?
- A normal distribution with mean = 0 and standard deviation = 1.
- Used in Z-score calculations.
13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- It states that the sampling distribution of the mean approaches a normal distribution as sample size increases.
- Important for hypothesis testing and confidence intervals.
14. How does the Central Limit Theorem relate to the normal distribution?
- It explains why normal distribution appears in many real-world situations, even when the original data isn’t normal.
15. What is the application of Z statistics in hypothesis testing?
Used when sample size is large (n > 30) and population variance is known.
Helps in determining statistical significance.
16. How do you calculate a Z-score, and what does it represent?

17. What are point estimates and interval estimates in statistics?
- Point estimate: A single value estimate of a population parameter.
- Interval estimate: A range of values within which the parameter is expected to lie (e.g., confidence intervals).
18. What is the significance of confidence intervals in statistical analysis?
- Provides a range of values to estimate an unknown population parameter with a certain level of confidence.
19. What is the relationship between a Z-score and a confidence interval?
- Z-score helps determine confidence intervals by specifying how far values can deviate from the mean.
20. How are Z-scores used to compare different distributions?

14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, value):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_score = (value - mean) / std_dev
    return z_score

def plot_standard_normal(z_score):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.plot(x, y, color='blue', label="Standard Normal Distribution")
    plt.axvline(z_score, color='red', linestyle='dashed', label=f"Z-score = {z_score:.2f}")

    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.title("Z-score Comparison with Standard Normal Distribution")
    plt.legend()
    plt.grid()
    plt.show()

# Example Usage
data = np.random.normal(50, 10, 1000)  # Sample dataset with mean=50, std=10
value = 60  # Data point to compare

z = calculate_z_score(data, value)
print(f"Z-score for {value}: {z:.2f}")

plot_standard_normal(z)
```

15.  Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    standard_error = population_std / np.sqrt(sample_size)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / standard_error

    # Calculate critical value for given alpha (two-tailed test)
    z_critical = norm.ppf(1 - alpha / 2)

    # Determine whether to reject the null hypothesis
    reject_null = abs(z_score) > z_critical

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.2f}")
    print(f"Critical Value: ±{z_critical:.2f}")
    print("Result: Reject Null Hypothesis" if reject_null else "Result: Fail to Reject Null Hypothesis")

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(50, 10, 30)  # Sample dataset (mean=50, std=10, size=30)
z_test(sample_data, population_mean=52, population_std=10, alpha=0.05)
```

16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy.stats import norm

def confidence_interval(sample, confidence=0.95):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Use sample standard deviation (ddof=1)
    sample_size = len(sample)
    standard_error = sample_std / np.sqrt(sample_size)

    # Compute margin of error
    z_critical = norm.ppf(1 - (1 - confidence) / 2)
    margin_of_error = z_critical * standard_error

    # Confidence Interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Confidence Interval ({int(confidence * 100)}%): ({lower_bound:.2f}, {upper_bound:.2f})")

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(50, 10, 30)  # Sample dataset (mean=50, std=10, size=30)
confidence_interval(sample_data, confidence=0.95)
```

19.  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lam=5, max_x=15):
    x = np.arange(0, max_x + 1)
    cdf_values = poisson.cdf(x, lam)

    plt.step(x, cdf_values, where="mid", color="blue", label=f"Poisson CDF (λ={lam})")
    plt.xlabel("X (Number of Events)")
    plt.ylabel("Cumulative Probability")
    plt.title("Poisson Distribution CDF")
    plt.legend()
    plt.grid()
    plt.show()

    print(f"CDF Values: {list(zip(x, np.round(cdf_values, 3)))}")

# Example Usage
plot_poisson_cdf(lam=5, max_x=15)
```

20. Simulate a random variable using a continuous uniform distribution and calculate its expected value.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

def simulate_uniform(a=0, b=10, size=1000):
    sample = np.random.uniform(a, b, size)
    expected_value = (a + b) / 2  # Expected value formula for uniform distribution

    # Plot the histogram
    plt.hist(sample, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black', label="Simulated Data")

    # Plot the theoretical uniform PDF
    x = np.linspace(a, b, 1000)
    y = uniform.pdf(x, a, b-a)
    plt.plot(x, y, 'r-', label="Uniform PDF")

    plt.xlabel("X values")
    plt.ylabel("Density")
    plt.title(f"Uniform Distribution U({a}, {b})")
    plt.legend()
    plt.grid()
    plt.show()

    print(f"Expected Value (Theoretical): {expected_value:.2f}")
    print(f"Sample Mean (Simulated): {np.mean(sample):.2f}")

# Example Usage
simulate_uniform(a=0, b=10, size=1000)
```

22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_range_and_iqr(data):
    data_range = np.max(data) - np.min(data)  # Range
    q1, q3 = np.percentile(data, [25, 75])  # First and third quartiles
    iqr = q3 - q1  # Interquartile Range

    print(f"Range: {data_range:.2f}")
    print(f"Interquartile Range (IQR): {iqr:.2f}")

    # Boxplot Visualization
    plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
    plt.xlabel("Values")
    plt.title("Boxplot of the Dataset")
    plt.grid()
    plt.show()

# Generate normal data
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10, Size=1000

# Calculate and display range & IQR
calculate_range_and_iqr(data)
```

23. Implement Z-score normalization on a dataset and visualize its transformation.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def z_score_normalization(data):
    normalized_data = zscore(data)  # Apply Z-score normalization

    print(f"Before Normalization: Mean = {np.mean(data):.2f}, Std Dev = {np.std(data, ddof=1):.2f}")
    print(f"After Normalization: Mean = {np.mean(normalized_data):.2f}, Std Dev = {np.std(normalized_data, ddof=1):.2f}")

    # Plot Histograms
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.hist(data, bins=30, alpha=0.6, color='blue', edgecolor='black')
    plt.title("Original Data Distribution")
    plt.xlabel("Values")
    plt.ylabel("Frequency")

    plt.subplot(1, 2, 2)
    plt.hist(normalized_data, bins=30, alpha=0.6, color='red', edgecolor='black')
    plt.title("Z-score Normalized Data")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()

# Generate a sample dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10, Size=1000

# Apply Z-score normalization and visualize
z_score_normalization(data)
```
